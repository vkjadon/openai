We can load the api key from `.env` file by importing `load_dotenv` method of `dotenv` package. We can install `dotenv` package as

> `pip install python-dotenv`

> `from dotenv import load_dotenv, find_dotenv`

`find_dotenv()` : This function searches up the directory tree from your current working directory until it locates the .env file.

`load_dotenv()` : This function reads the variables from the found .env file and loads them into your Python process's environment variables (specifically, os.environ).

```python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())

client = OpenAI()
```

In case you are running the code on Google Colab, add your "OPENAI_API_KEY" key in the `Secrets` of the colab environment and call instantiate the `OpenAI` object as below

```python
from openai import OpenAI
from google.colab import userdata
client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

It is stored server-side on OpenAI’s infrastructure

```python
conversation = client.conversations.create()
conversation_id = conversation.id

print("Conversation ID:", conversation_id)
```

client.conversations.create() stored server-side on OpenAI's infrastructure. Conversation still exists on OpenAI servers
(as long as you keep the conversation_id), If Colab kernel crashes, Notebook restarts, Browser closes

```python
# Turn 1
r1 = client.responses.create(
    model="gpt-4.1-mini",
    conversation=conversation_id,
    input="Explain PID control in simple words"
)
print("Assistant:", r1.output_text)
```

```python
# Turn 2
r2 = client.responses.create(
    model="gpt-4.1-mini",
    conversation=conversation_id,
    input="Explain only the integral term"
)

print("Assistant:", r2.output_text)
```

```python
analysis_response = client.responses.create(
    model="gpt-4.1-mini",
    conversation=conversation_id,
    input="""
Based on our previous discussion,
1. Identify what topic the student is currently focusing on
2. Check if the student is confused or progressing well
3. Suggest the next logical concept to explain
"""
)

print(analysis_response.output_text)
```

```python
conversation_data = client.conversations.retrieve(conversation_id)
print(conversation_data)
```

```python
items = client.conversations.items.list(conversation_id, limit=10)
print(items.data)
```

```python
for item in items:
    print(f"{item.role.upper()}: {item.content[0].text}")
```
